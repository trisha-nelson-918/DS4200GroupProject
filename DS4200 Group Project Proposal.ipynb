## Introductory Visualizations
### Part 1: Scatterplot

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('movies_cleaned.csv')
df.head()
```

```python
# finding the data across popularity scores
df['popularity'].describe()
```

The average popularity score is around 20. There are some outliers that are heavily skewing our data (values greater than 60), so I will not be taking those into account. I will only be taking values that are within 1 standard deviation from our mean (80). 

```python
# setting new parameters for box office 
df = df[df['revenue'] > 0]
df = df[df['popularity'] >= 1]
df = df[df['popularity'] < 80]
```

```python
# scatter plot of box office revenue and audience popularity scores
plt.scatter(df['popularity'],df['revenue'])
plt.yscale('log')
plt.xlabel('Audience Popularity Score')
plt.ylabel('Box Office Revenue')
plt.title('Box Office Revenue vs Audience Popularity Scores')
```

```python
# backup histogram
plt.hist(df['vote_average'], bins = 10)
plt.title('Frequency of Average Viewer Ratings')
plt.ylabel('Frequency')
plt.xlabel('Average Rating')
```
